```python
import phylustrator as ph

# Load and Plot
data = ph.parse_alerax("../examples/data/alerax")
v_drawer = ph.VerticalTreeDrawer(data.species_tree, style=ph.TreeStyle(width=800, height=800))

# Plot only the top 5% of transfers
v_drawer.add_transfer_links(data.transfers, arrows=True, top_percentile=0.001)
v_drawer.add_leaf_names()
v_drawer.draw()
v_drawer.d
```
