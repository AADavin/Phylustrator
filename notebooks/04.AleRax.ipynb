```python
import phylustrator as ph

# 1. Parse AleRax data
data = ph.parse_alerax("../examples/data/alerax")

# 2. Setup Vertical Figure
v_style = ph.TreeStyle(width=600, height=800)
v_drawer = ph.VerticalTreeDrawer(data.species_tree, style=v_style)
v_drawer.plot_transfers(data.transfers, filter_below=0.9, arc_intensity=40, curve_type="S", stroke_width=0.9)
v_drawer.add_leaf_names()
v_drawer.draw()

# Display them
print("Vertical Layout:")
display(v_drawer.d)
```

```python
import phylustrator as ph

# 1. Parse AleRax data
data = ph.parse_alerax("../examples/data/alerax")

# --- VERTICAL DRAWER ---
print("Rendering Vertical Layout...")
v_style = ph.TreeStyle(width=600, height=800)
v_drawer = ph.VerticalTreeDrawer(data.species_tree, style=v_style)

# Plot transfers with S-curve or C-curve
v_drawer.plot_transfers(
    data.transfers, 
    filter_below=0.9, 
    arc_intensity=40, 
    curve_type="S",
    use_thickness=True,
    opacity=0.7,
    stroke_width=1
)
v_drawer.add_leaf_names()
v_drawer.draw()

v_drawer.add_transfer_legend(low=0.1, high=1.0)
display(v_drawer.d)
```

```python
# --- RADIAL DRAWER ---
print("\nRendering Radial Layout...")
r_style = ph.TreeStyle(width=800, height=800)
r_drawer = ph.RadialTreeDrawer(data.species_tree, style=r_style)

# Radial automatically handles layout internally now
r_drawer.plot_transfers(
    data.transfers, 
    filter_below=0.1, 
    arc_intensity=50, 
    curve_type="C"
)
r_drawer.add_leaf_names()
r_drawer.draw()

# Add legend
r_drawer.add_transfer_legend(low=0.1, high=1.0)
display(r_drawer.d)
```
